```python
you are given with array of numbers.you have to find whether array is beautiful or not. A beautiful array is an array whose sum of all numbers is divisible by 2, 3 and 5

Input Description:
You are given a number ‘n’ denoting the size of array.Next line contains n space separated numbers.

Output Description:
Print 1 if array is beautiful and 0 if it is not

Sample Input :
5
5 25 35 -5 30
Sample Output :
1
```

```python
num = int(input())

a = [int(x) for x in input().split()]
s = 0

for i in range(0,num):
    s = s + a[i]

if s % 2 == 0 and s % 3 == 0 and s % 5 == 0:
    print(1)
else:
    print(0)

```

### Given a string 'S' print the sum of weight of the String. A weight of character is defined as the ASCII value of corresponding character. Input Description: You are given a string ‘s’   Output Description: Print weight : Sample Input : abc Sample Output : 294

```python
s = input()
su = 0
str_len = len(s)

for i in range(0,str_len):
    su = su + ord(s[i])
    
print(su)
    
```
