```python
from matplotlib import pyplot as plt
import pandas as pd
```

bar chart

```python
plt.bar([0.25,1.35,2.25,3.25,4.25], [30,40,50,60,70],label = "male", color = 'c', width = 0.5)
plt.bar([0.75,1.75,2.75,3.75,4.75], [50,60,70,80,90],label = "female", color = 'g', width = 0.5)
plt.legend()
plt.show()
```

scatter plot

```python
x1 = [1.5,2.5,3.5,4.5,6.5]
y1 = [8.5,6,56,42,892]

plt.scatter(x1,y1)
plt.show()
```

pie chart

```python
runs = [100,127,100,61]
players = ['rahul','rohit','virat', 'pujara']
plt.pie(runs, labels = players, startangle = 90,shadow = True, explode = (0,0.1,0,0), autopct = '%1.1f%%')
plt.show()
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

strip plot

```python
sns.stripplot(x = 'species', y = 'sepal_length', data = df)
plt.show()
```

violin plots

```python
sns.violinplot(x = 'species', y = 'sepal_length', data = df, splkit = True)
plt.show()
```

swarm plot

```python
sns.swarmplot(x = 'species', y = 'sepal_length', data = df)
plt.show()
```

box plot

```python
sns.boxplot(x = 'species', y = 'sepal_width', data = df)
plt.show()
```
